## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#Find number of Unique Items, Average Price, Number of Purchases and Total Revenue
unique_items = len(original_df['Item ID'].unique())
average_price = original_df['Price'].mean().round(decimals=2)
total_purchases = original_df['Purchase ID'].count()
total_revenue = original_df['Price'].sum()

#Display them as summary
purchase_summary = pd.DataFrame({"Number of Unique Items": unique_items, "Average Price": average_price, 
                                 "Number of Purchases": total_purchases, "Total Revenue": total_revenue}, index = [0])

#Format values for proper display
purchase_summary["Average Price"] = purchase_summary["Average Price"].map("${:.2f}".format)
purchase_summary["Total Revenue"] = purchase_summary["Total Revenue"].map("${:,.2f}".format)

purchase_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Create an index for unique players
unique_sn = original_df.drop_duplicates(subset = "SN")
by_gender = original_df.groupby("Gender")

#Find the number and percentage of Players by Gender
count_gender = unique_sn["Gender"].value_counts()
gender_pct = count_gender / unique_sn["Gender"].count() * 100

#Display them as summary
gender_summary = pd.DataFrame({"Total Count": count_gender, "Percentage of Players": gender_pct})

#Format values for proper display
gender_summary["Percentage of Players"] = gender_summary["Percentage of Players"].map("{:.2f}%".format)

gender_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Group age bins, age groups and assign them to unique_sn 
bins = [0,9,14,19,24,29,34,39,100]
age_groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
unique_sn["Age Range"] = pd.cut(unique_sn["Age"], bins, labels = age_groups)

#Count number and percentage of players by age
purchase_count_age = unique_sn["Age Range"].value_counts()
pct_player_age = (purchase_count_age / unique_sn["SN"].count()) * 100

#Create age summary and sort by index
age_summary_df = pd.DataFrame({"Total Count" : purchase_count_age,
                               "Percentage of Players" : pct_player_age})
age_summary_df = age_summary_df.sort_index()

#Format the currency values
age_summary_df["Percentage of Players"] = age_summary_df["Percentage of Players"].map("{:.2f}%".format)

age_summary_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Categorize the total players using the age bins 
original_df["Age Range"] = pd.cut(original_df["Age"], bins, labels = age_groups)
purchase_age_count = original_df["Age Range"].value_counts()

#Calculate average purchase price, total purchase value and average purchase per person based on age bins
average_price = original_df.groupby("Age Range")["Price"].mean()
total_purchase_age = original_df.groupby("Age Range")["Price"].sum()
average_purchase_age = total_purchase_age / purchase_count_age


#Display them as summary
purchase_summary_age = pd.DataFrame({"Purchase Count" : purchase_age_count,
                                     "Average Purchase Price" : average_price,
                                     "Total Purchase Value" : total_purchase_age,
                                     "Avg Total Purchase per Person" : average_purchase_age})

#Format the currency values
purchase_summary_age["Average Purchase Price"] = purchase_summary_age["Average Purchase Price"].map("${:.2f}".format)
purchase_summary_age["Total Purchase Value"] = purchase_summary_age["Total Purchase Value"].map("${:.2f}".format)
purchase_summary_age["Avg Total Purchase per Person"] = purchase_summary_age["Avg Total Purchase per Person"].map("${:.2f}".format)

#Rename the axis to show data label Age Ranges
purchase_summary_age = purchase_summary_age.rename_axis("Age Ranges")

purchase_summary_age
```
